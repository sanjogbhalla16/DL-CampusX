```python
import tensorflow
from tensorflow import keras
from keras import Sequential
from keras.layers import Dense,Flatten,SimpleRNN
```

```python
model = Sequential()

model.add(SimpleRNN(3, input_shape=(4,5)))
model.add(Dense(1, activation='sigmoid'))

model.summary()
```

```python
model.get_weights()
```

```python
print(model.get_weights()[0].shape)
model.get_weights()[0]
```

```python
print(model.get_weights()[1].shape) #feedback connection
model.get_weights()[1]
```

```python
print(model.get_weights()[2].shape) #bias for the hidden layer
model.get_weights()[2]
```

```python
print(model.get_weights()[3].shape) # weights for output layer
model.get_weights()[3]
```

```python
print(model.get_weights()[4].shape) #bias for the output layer
model.get_weights()[4]
```
